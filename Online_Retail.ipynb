#Online Retail

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
url="https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Online_Retail/Online_Retail.csv"
online_rt=pd.read_csv(url,sep=',',encoding = 'latin1')
print(online_rt)
```

```
# Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK
print("Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK:")
r=online_rt.groupby('Country').sum()
print(r)

r.sum()[1:11]

r.plot(kind='bar')
plt.xlabel('Quantity')
plt.ylabel('Country')
plt.title("10 countries by'Quantity' ordered except UK")
plt.show()
```
