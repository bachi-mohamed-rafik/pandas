# US Crime Rate

```
import pandas as pd
url="https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
crime=pd.read_csv(url,sep=',')
print(crime)
```

```
# What is the type of the columns?
print("What is the type of the columns:")
print(crime.dtypes)

# Convert the type of the column Year to datetime64
print('Convert the type of the column Year to datetime64 :')
print(pd.to_datetime(crime['Year']))
```

```
# Set the Year column as the index of the dataframe
print("Set the Year column as the index of the dataframe :")
crime.set_index('Year', drop = True)

# Delete the Total column
print("Delete the Total column :")
del crime['Total']
print(crime.columns)
```
